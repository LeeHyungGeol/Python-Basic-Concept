```python
f = open('C:/python_workspace/common/test2.txt', 'r') # r은 생략 가능하다.

while True:  # 무한 루프
    line = f.readline()
    # 문자열(String)은 내용이 있으면 True, 없으면 False이다.
    if not line: break
    print(line, end='')  # 맨 마지막 문자(개행 문자)를 지운다.
    
f.close()        
```

```python
# csv 파일 : 쉼표를 구분자로 갖고 있는 파일 형태

nums = range(10)  # 0 ~ 9 까지
print(list(nums))
cnt = len(nums)  # len(리스트) : 리스트의 갯수

output_file = 'result.csv'

f = open(output_file, 'a')
for i in range(cnt):
    if i < (cnt-1):
        f.write(str(nums[i]) + ',')
    else:
        f.write(str(nums[i]) + '\n')
f.close()
print('저장되었습니다')
```

```python
# 자료출처 : 고공데이터포털
# 서울교통공사_승차인원_2019_1월~7월
# https://www.data.go.kr/dataset/15003169/fileData.do

with open('C:/python_workspace/data/traffic/data.csv') as f:
    rows = f.readlines() #파일 전체의 내용을 list로 저장
    #print(len(rows))
    #print(rows)
    
    a ='' #파일 전체 내용을 저장하기 위한 변수
    for row in rows:   # 전체의 행에서 하나의 행씩 읽어온다
        a += row.strip()   # 각 행의 좌우 공백을 제거
    number = input('찾을 숫자를 입력하세요 :')
    
    # A in B : B에 A가 있으면 True 없으면 False
    if number in a:
        print('숫자를 찾았습니다.')
        tot = a.count(number)
        print('찾은 횟수 : ',tot)
    else:
        print(number + '숫자를 찾지 못했습니다.')
```
